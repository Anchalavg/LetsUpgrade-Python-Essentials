```
# Filter
def is_prime(n): 
    if n > 1: 
       for i in range(2,n): 
           if (n % i) == 0: 
              break
       else:
           print(n,"is prime")
```

```
is_prime = list (filter(is_prime,range(1,2500)))
print (is_prime)
```

#Finding subsetlist

```
 m = [1,5,6,4,1,2,3,5]
n = [1,1,5]
 
def check(a,b):
    for i in a:
        for j in b:
            if(len(b)>=0 and j==i):
               b.pop(0)
            else:
               break
    return b
if len(check(m,n))==0:
   print("its a match")
else:
   print("its gone")
```
